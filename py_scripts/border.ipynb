```python
import matplotlib.pyplot as plt

import math
import numpy as np

wide = np.radians(55/2)
g = 6.

x=np.linspace(-np.pi/g, np.pi/g, 10, endpoint = True)
print(np.pi)
print("ширина обзора в радианах = {}".format(wide))
B1 = -np.tan(x+wide)
C1 = -0.2*np.sqrt(1+np.square(np.tan(x+wide)))

B2 = -np.tan(x-wide)
C2 = 0.2*np.sqrt(1+np.square(np.tan(x-wide)))

fig = plt.figure(1)

gr1 = fig.add_subplot(2, 2, 1)
gr1.set(title = 'заисимость B1 от Thi')
gr1.plot(x, B1)
gr1.scatter(x, B1, color='orange')

gr2 = fig.add_subplot(2, 2, 2)
gr2.set(title = 'заисимость C1 от Thi')
gr2.plot(x, C1, 'g--')
gr2.scatter(x, C1, color='orange')

gr3 = fig.add_subplot(2, 2, 3)
gr3.set(title = 'заисимость B2')
gr3.plot(x, B2, 'r--')
gr3.scatter(x, B2, color='orange')

gr3 = fig.add_subplot(2, 2, 4)
gr3.set(title = 'заисимость C2')
gr3.plot(x, C2, 'r--')
gr3.scatter(x, C2, color='orange')

plt.rcParams['figure.figsize'] = (plt.rcParamsDefault['figure.figsize'][0]*3, plt.rcParamsDefault['figure.figsize'][0]*3)

plt.show()

x0 = np.linspace(-1, 1, num = 100, endpoint = True)

k=1
fig0=plt.figure(2, figsize=(20, 20))
grf = fig0.add_subplot(1, 1, 1)
grf.set(title = ' Диапазон угла отклонения:  {} deg - {} deg'.format(np.rad2deg(-np.pi/g), np.rad2deg(np.pi/g)))

for fi in x:
    B01 = math.tan(fi+wide)
    C01 = 0.2*math.sqrt(1+(math.tan(fi+wide))**2)
    B02 = math.tan(fi-wide)
    C02 = 0.2*math.sqrt(1+(math.tan(fi-wide))**2)

    y01 = (x0-C01)/B01
    y02 = (x0+C02)/B02

    fig1 = plt.figure(3, figsize=(20,140))

    grk = fig1.add_subplot(20, 2, k)
    grk.set(title = ' Thi = {} deg'.format(np.rad2deg(fi)))
    k = k+1

    grk.plot(x0, y01, 'r--')
    grk.plot(x0, y02, 'g--')
    grk.set_xlabel(xlabel='X0')
    grk.set_ylabel("Y0")
    grk.set_ylim([0, 2])

    grf.plot(x0, y01, 'r--')
    grf.plot(x0, y02, 'g--')
    grf.set_xlabel(xlabel='X0')
    grf.set_ylabel("Y0")
    grf.set_ylim([0, 3])

plt.rcParams['figure.figsize'] = (plt.rcParamsDefault['figure.figsize'][0], plt.rcParamsDefault['figure.figsize'][0])
plt.show()

```
